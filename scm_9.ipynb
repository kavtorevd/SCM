# **Задача 9: Как растет число обусловленности для случайной матрицы?**

## Выполнил Кавторев Дмитрий Константинович

Цель данного упражнения – экспериментально выяснить зависимость числа
обусловленности $\mbox{cond}_1$ от порядка случайной матрицы. Пусть $A_n$ –
матрица порядка $n$ со случайными элементами, распределенными по
нормальному закону с мат ожиданием $0$ и средним квадратическим отклонением
$1$. Экспериментальным путем необходимо выяснить, что в среднем
$\mbox{cond}_1 A_n$ растет как показательная функция от $n$, а именно, что
для большинства матриц
$$
c_1 n^p \le \mbox{cond}_1 A_n \le c_2 n^p.
$$

Подключение библиотек:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Устанавливаем число экспериментов и значение ранга матрицы:



```python
n_min = 1
n_max = 100
num_matrices = 100
```

Вручную подберем значения для показательных функций:

```python
p = 2
c1 = 1
c2 = 10
```

Вычисляем показательные функции

```python
ns = np.arange(n_min, n_max+1)
c1_n_p = c1 * ns**p
c2_n_p = c2 * ns**p
```

Вычисляем значение числа обсуловленности матриц и показательных функций:

```python
conds = []
for n in ns:
    matrices = np.random.normal(0, 1, size=(num_matrices, n, n))
    conds_n = [np.linalg.cond(matrix, 1) for matrix in matrices]
    conds.append(np.mean(conds_n))
```

Построим график чисел обусловленности и показательных функций:

```python
fig, ax = plt.subplots()
ax.loglog(ns, conds, '.b', markersize=5, label='cond_1*A_n')
ax.loglog(ns, c1_n_p, 'r', label='c1*n^p')
ax.loglog(ns, c2_n_p, 'g', label='c2*n^p')
ax.legend(loc='upper left')
ax.set_xlabel('n')
ax.set_ylabel('cond_1')
plt.show()
```

### Вывод: я провел эксперимент над несколькими случайными матрицами. Было найдено число обусловленности, а также был построен график для каждой из матриц. Подобраны значения $p, c_1, c_2$ и нарисованы графики функций $c_1n^P, c_2n^P.$


### $$
c_1 n^p \le \mbox{cond}_1 A_n \le c_2 n^p
$$
### Таким образом, для большинства матриц это выражение верно.
